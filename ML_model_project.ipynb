# set dependent and independent variable

```python
y=ds['Sold']
```

```python
x=ds.loc[:,ds.columns!='Sold']
```

```python
x.head()
```

```python
y.head()
```

# split the data into train and test

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

# Build and train the model based on test data

```python
from sklearn.linear_model import LogisticRegression
```

```python
#create object and fit

ob=LogisticRegression()
```

```python
ob.fit(x_train,y_train)
```

```python
#prediction for training data and confusion matrix
y_train_p=ob.predict(x_train)
y_train_p
```

```python
ob.predict_proba(x_train)
```

```python
from sklearn.metrics import confusion_matrix,precision_score
```

```python
confusion_matrix(y_train,y_train_p)
```

```python
precision_score(y_train,y_train_p)
```

# for test data

```python
y_test_p=ob.predict(x_test)
```

```python
y_test_p
```

```python
confusion_matrix(y_test,y_test_p)
```
